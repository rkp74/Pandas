# Merge and concat

Merge

```python
import pandas as pd
```

```python
df4 = pd.DataFrame({"Roll":[1,2,3,4,5] , "Phy_Marks" : [8,9,6,7,3]})
df5 = pd.DataFrame({"Roll":[1,2,3,4,5] , "Math_Marks" : [6,8,7,5,6]})

df6 = pd.merge(df5,df4,on="Roll")
df6
```

```python
df7 = pd.DataFrame({"Roll":[1,2,3,4,5] , "Phy_Marks" : [8,9,6,7,3]})
df8 = pd.DataFrame({"Roll":[1,2,3,4,6] , "Math_Marks" : [6,8,7,5,6]})
# only common Roll got merged
df9 = pd.merge(df7,df8,on="Roll")
df9
```

```python
df10 = pd.DataFrame({"Roll":[1,2,3,4,5] , "Phy_Marks" : [8,9,6,7,3]})
df11 = pd.DataFrame({"Roll":[1,2,3,4,6] , "Math_Marks" : [6,8,7,5,6]})
# only common Roll got merged
df12 = pd.merge(df10,df11,how ="inner")
df12
```

```python
df13 = pd.DataFrame({"Roll":[1,2,3,4,5] , "Phy_Marks" : [8,9,6,7,3]})
df14 = pd.DataFrame({"Roll":[1,2,3,4,6] , "Math_Marks" : [6,8,7,5,6]})
# only common Roll got merged
df15 = pd.merge(df13,df14,how ="left")
df15
```

```python
df16 = pd.DataFrame({"Roll":[1,2,3,4,5] , "Phy_Marks" : [8,9,6,7,3]})
df17 = pd.DataFrame({"Roll":[1,2,3,4,6] , "Math_Marks" : [6,8,7,5,6]})
# only common Roll got merged
df18 = pd.merge(df16,df17,how ="right")
df18
```

```python
df19 = pd.DataFrame({"Roll":[1,2,3,4,5] , "Phy_Marks" : [8,9,6,7,3]})
df20 = pd.DataFrame({"Roll":[1,2,3,4,6] , "Math_Marks" : [6,8,7,5,6]})
# only common Roll got merged
df21 = pd.merge(df19,df20,how ="outer", indicator = True)
df21
```

```python
df22 = pd.DataFrame({"Roll":[1,2,3,4,5] , "Phy_Marks" : [8,9,6,7,3]})
df23 = pd.DataFrame({"Roll":[1,2,3,4,6] , "Phy_Marks" : [6,8,7,5,6]})
# only common Roll got merged
df24 = pd.merge(df22,df23,left_index = True , right_index = True, suffixes=("name" , "id"))
df24
```

Pivot

```python
df50 = pd.DataFrame({"days" : [1,2,3,4,5,6] , "student_name":["a" , "b" ,"c" ,"a" , "b" ,"c"] , "eng" : [10,12,14,15,16,12] , "maths" : [17,18,19,13,14,16]})
df50
```

```python
df50.pivot(index = "days" , columns = "student_name")
```

```python
df51 = pd.DataFrame({"days" : [1,2,3,4,5,6] , "student_name":["a" , "b" ,"a" ,"a" , "b" ,"b"] , "eng" : [10,12,14,15,16,12] , "maths" : [17,18,19,13,14,16]})

df51.pivot(index = "days" , columns = "student_name" , values = "eng")
```

```python
df52 = pd.DataFrame({"days" : [1,1,2,1,2,1] , "student_name":["a" , "b" ,"a" ,"a" , "b" ,"a"] , "eng" : [10,12,14,15,16,12] , "maths" : [17,18,19,13,14,16]})
df52
```

```python
df52.pivot_table(index = "student_name" , columns = "days" ,
                aggfunc = "mean")
```

```python
df52.pivot_table(index = "student_name" , columns = "days" ,
                aggfunc = "sum")
```

```python
df52.pivot_table(index = "student_name" , columns = "days" ,
                aggfunc = "sum" , margins = True)
```

```python
df52.pivot_table(index = "student_name" , columns = "days" ,
                aggfunc = "mean" , margins = True)
```
